```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
```

```python
#pulling in data
enrolled_df = pd.read_csv("C:\\Users\\ekims\\Documents\\NSS\\DA14\\capstone\\capstone\\data\\enrolled_pop.csv")
```

```python
#creating sorted senior gap
senior_gap = enrolled_df[['Geographic Area Name','County', 'State', 'senior_pop', 'Senior Pop Enrolled', 'Senior Medicaid Coverage Rate',
                          'State_Enrollment_Senior', 'State_Pop_Senior', 'Senior Gap']].sort_values(by='Senior Gap', ascending=False).reset_index()
top_5_senior = senior_gap.head(5)
```

```python
perc_enrolled_nobronx_df = pd.read_csv("C:\\Users\\ekims\\Documents\\NSS\\DA14\\capstone\\capstone\\data\\perc_enrolled_nobronx.csv")
perc_enrolled_nobronx_df.info()
```

```python
bar_height = 0.35
y_pos = np.arange(len(perc_enrolled_nobronx_df))

plt.figure(figsize=(10, 6))

plt.barh(y_pos - bar_height/2, perc_enrolled_nobronx_df['E_Total_pop'], height=bar_height, label='Total Pop', color='cornflowerblue')
plt.barh(y_pos + bar_height/2, perc_enrolled_nobronx_df['Total Pop Enrolled'], height=bar_height, label='Medicaid Enrolled', color='lightsalmon')

plt.yticks(y_pos, perc_enrolled_nobronx_df['County'])

plt.xlabel('Population')
plt.title('Total Population vs. Medicaid Enrolled')
plt.legend()
plt.gca().invert_yaxis()
plt.grid(True, axis='x')
plt.tight_layout()
plt.savefig('pop_vs_enrolled_cluster', bbox_inches='tight')
plt.show()
```
