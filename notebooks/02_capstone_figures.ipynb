```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#pulling in data
enrolled_df = pd.read_csv("C:\\Users\\ekims\\Documents\\NSS\\DA14\\capstone\\capstone\\data\\enrolled_pop.csv")
```

```python
enrolled_df.info()
```

```python
#creating sorted senior gap
senior_gap = enrolled_df[['Geographic Area Name','County', 'State', 'senior_pop', 'Senior Pop Enrolled', 'Senior Medicaid Coverage Rate',
                          'State_Enrollment_Senior', 'State_Pop_Senior', 'Senior Gap']].sort_values(by='Senior Gap', ascending=False).reset_index()
top_5_senior = senior_gap.head(5)
```

```python
plt.barh(top_5_senior['Geographic Area Name'], top_5_senior['Senior Gap'], color="skyblue")

plt.xlabel('Eligible Seniors Not Enrolled')
plt.ylabel('County, State')
plt.title('Senior Enrollment Gap, Top 5 Counties')
plt.xticks(rotation = 70)
plt.gca().invert_yaxis()
plt.savefig('top_5_senior', bbox_inches='tight')
```

```python
adult_gap = enrolled_df[['Geographic Area Name', 'County', 'State', 'adult_pop', 'Adult Pop Enrolled', 'Adult Medicaid Coverage Rate',
                          'State_Enrollment_Adult', 'State_Pop_Adult', 'Adult Gap']].sort_values(by='Adult Gap', ascending=False).reset_index()
top_5_adult = adult_gap.head(5)
```

```python
plt.barh(top_5_adult['Geographic Area Name'], top_5_adult['Adult Gap'], color='steelblue')

plt.xlabel('Eligible Adults Not Enrolled')
plt.ylabel('County, State')
plt.title('Top 5 Counties, Adult Enrollment Gap (based on income)')
plt.gca().invert_yaxis()
plt.xticks(rotation = 70)
plt.savefig('top_5_adult', bbox_inches='tight')
```
